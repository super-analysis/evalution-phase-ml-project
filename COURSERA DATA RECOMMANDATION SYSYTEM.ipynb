```python
#removing spaces
data['course_name'] = data['course_name'].str.replace(' ',',')
data['course_name'] = data['course_name'].str.replace(',,',',')
data['course_name'] = data['course_name'].str.replace(':','')
data['skills'] = data['skills'].str.replace(' ',',')
data['skills'] = data['skills'].str.replace(',,',',')
data['skills'] = data['skills'].str.replace('_','')
data['skills'] = data['skills'].str.replace(':','')
data['skills'] = data['skills'].str.replace('(','')
data['skills'] = data['skills'].str.replace(')','')
```

# feature engineering

```python
new_df = data[['course_name','tags']]
```

```python
new_df['tags'] = data['tags'].str.replace(',',' ')
```

```python
new_df['course_name'] = data['course_name'].str.replace(',',' ')
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
#lower casing
```

```python
new_df.head()
```

# text vectraization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

# stemming process

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)#applying stemming tag columns
```

# similarity measures

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

# recommendation function

# evalution model

```python
def recommend(course):
    course_index = new_df[new_df['course_name'] == course].index[0]
    distances = similarity[course_index]
    course_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:7]
    
    for i in course_list:
        print(new_df.iloc[i[0]].course_name)
```

```python
recommend('IBM Applied DevOps Engineering')
```
